```python
mainroad_dummie = pd.get_dummies(df['mainroad'], drop_first=True, dtype=int)

guestroom_dummie = pd.get_dummies(df['guestroom'], drop_first=True, dtype=int)

basement_dummie = pd.get_dummies(df['basement'], drop_first=True, dtype=int)

hotwater_dummie = pd.get_dummies(df['hotwaterheating'], drop_first=True, dtype=int)

aircon_dummie = pd.get_dummies(df['airconditioning'], drop_first=True, dtype=int)

prefarea_dummie = pd.get_dummies(df['prefarea'], drop_first=True, dtype=int)

furnish_dummie = pd.get_dummies(df['furnishingstatus'], drop_first=True, dtype=int)

df = df.drop(df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']], axis=1)

df = pd.concat([df,mainroad_dummie, guestroom_dummie, basement_dummie, hotwater_dummie, aircon_dummie, prefarea_dummie, furnish_dummie],axis=1)
```

```python
# Splitting the data

X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Scaling the data

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Linear regression model

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train,y_train)

predictions_lr = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test,predictions_lr))
print('MSE: ', metrics.mean_squared_error(y_test,predictions_lr))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test,predictions_lr)))
```

```python
sns.distplot((y_test-predictions_lr),bins=50);
```

```python
lr.score(X_train,y_train)
```

```python
# Random Forest model

from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()

rfr.fit(X_train_scaled, y_train)

predictions_rfr = rfr.predict(X_test_scaled)
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test,predictions_rfr))
print('MSE: ', metrics.mean_squared_error(y_test,predictions_rfr))
print('RMSE: ', np.sqrt(metrics.mean_absolute_error(y_test,predictions_rfr)))
```

```python
# It seems Random Forest did a much better job in predicting the correct price.

rfr.score(X_train_scaled,y_train)
```
